```python
kepler_df = kepler_df.dropna(axis='columns', how='all')
kepler_df = kepler_df.dropna()

kepler_df.head()
```

```python
kepler_df = kepler_df[['koi_disposition',
        'koi_score', 'koi_fpflag_nt', 'koi_fpflag_ss',
       'koi_fpflag_co', 'koi_fpflag_ec', 'koi_period', 'koi_period_err1',
       'koi_period_err2', 'koi_time0bk', 'koi_time0bk_err1',
       'koi_time0bk_err2', 'koi_impact', 'koi_impact_err1', 'koi_impact_err2',
       'koi_duration', 'koi_duration_err1', 'koi_duration_err2', 'koi_depth',
       'koi_depth_err1', 'koi_depth_err2', 'koi_prad', 'koi_prad_err1',
       'koi_prad_err2', 'koi_teq', 'koi_insol', 'koi_insol_err1',
       'koi_insol_err2', 'koi_model_snr', 'koi_tce_plnt_num',
        'koi_steff', 'koi_steff_err1', 'koi_steff_err2',
       'koi_slogg', 'koi_slogg_err1', 'koi_slogg_err2', 'koi_srad',
       'koi_srad_err1', 'koi_srad_err2', 'ra', 'dec', 'koi_kepmag']]
```

```python
selected_features = kepler_df.drop('koi_disposition', axis=1)
selected_features
```

```python
X = selected_features
y = kepler_df["koi_disposition"].values

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

```python
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=41))
model.add(Dense(units=2, activation='softmax'))
model.summary()
```

```python

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=3
)
```

```python
model.save("deep_learning_model.h5")
```

```python
from tensorflow.keras.models import load_model
model = load_model("deep_learning_model.h5")
```

```python
model_loss, model_accuracy = model.evaluate(X_train_scaled, y_train_categorical, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
